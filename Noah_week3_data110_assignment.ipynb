```python
def f(x): return x**2
def g(x): return x**3
def h(x): return x**4
def i(x): return x**5

x = np.linspace (-2, 2, 400)
y_f = f(x)
y_g = g(x)
y_h = h(x)
y_i = i(x)
plt.plot(x, y_f, label=r'f(x) = x^2', color='red', linestyle='-',)
plt.plot(x, y_g, label=r'g(x) = x^3', color='g', linestyle='--',)
plt.plot(x, y_h, label=r'h(x) = x^4', color='y', linestyle=':',)
plt.plot(x, y_i, label=r'i(x) = x^5', color='b', linestyle='-.',)
plt.title('Plot of f(x), g(x), h(x), i(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def f(x): return x*np.sin(x)

x = np.linspace(0, 4*np.pi, 100)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'line', linestyle='-', color='g', linewidth=1)
plt.scatter(x, y, s=25, label='points', color='b', edgecolor='y', linewidth=1)

plt.xlabel('X')
plt.ylabel('f(x)')
plt.title('xsin(x)')

plt.legend()
plt.show()
```

```python
def f(x):return x * np.cos(x)
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = f(x)
plt.plot(x, y, label=r'$f(x) = x \cos(x)$', color='blue', linestyle='-', linewidth=2)
x_scatter = np.arange(-2 * np.pi, 2 * np.pi + np.pi/4, np.pi/4)
y_scatter = f(x_scatter)
plt.scatter(x_scatter, y_scatter, color='red', marker='o', label='Scatter points', zorder=5)
plt.title(r'Plot of $f(x) = x \cos(x)$ with Scatter Points', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
