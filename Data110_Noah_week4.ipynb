```python
import matplotlib.pyplot as plt
import numpy as np
```

## Task 1
#Basic Bar Chart for Crime Rates in 2023


```python
crime_types = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
crime_rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
plt.figure(figsize=(10, 6))
bars = plt.bar(crime_types, crime_rates_2023, color=['green'])

plt.title('Crime Rates per 100,000 People in 2023', fontsize=14)
plt.xlabel('Crime Type', fontsize=12)
plt.ylabel('Crime Rate (per 100,000)', fontsize=12)
```

```python
crime_types = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
percentage_change = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
plt.figure(figsize=(10, 6))
bars = plt.bar(crime_types, percentage_change, color=['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink'])

plt.title('Percentage Change in Crime Rates', fontsize=14)
plt.xlabel('Crime Type', fontsize=12)
plt.ylabel('Percentage Change', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
bars = plt.barh(crime_types, percentage_change, color=plt.cm.Paired(range(len(crime_types))))
plt.grid(True, axis='x', linestyle='-', alpha=0.6)
plt.title('Percentage Change in Crime Rates (2022-2023)', fontsize=14)
plt.xlabel('Percentage Change (%)', fontsize=12)
plt.ylabel('Crime Type', fontsize=12)
for bar in bars:plt.text(bar.get_width() - 0.5, bar.get_y() + bar.get_height() / 2,
              f'{bar.get_width():.1f}%', va='center', ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```
