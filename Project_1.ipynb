```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data
yearly_data = {
"Year": [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
"Gun Deaths": [500, 600, 550, 700, 720, 800, 850, 900, 880, 950]
}
```

```python
df_year = pd.DataFrame(yearly_data)

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_year, x="Year", y="Gun Deaths", marker='o', linewidth=2)
plt.title("Figure 1: Trend in Gun Deaths by Year (2010â€“2019)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Gun Deaths")
plt.tight_layout()
plt.show()
```

# Visualization 2: Gun Deaths by Age Group

Youth from 18 to 34 years old the highest risk category regarding gun violence incidents, according to the data. Gun-related fatalities claim the most lives in people aged 25-34 than in 18-24-year-olds, respectively. The yearly gun death count exceeds 1,400 fatalities among these three specific age groups. The demographic group needs special attention because they face heightened gun violence risks, which require improved mental health services alongside gang violence prevention.


# Deaths by Age Group:

```python
age_data = {
"Age Group": ["0-17", "18-24", "25-34", "35-44", "45-54", "55-64", "65+"],
"Gun Deaths": [1400, 1480, 1500, 700, 720, 800, 850]
}
```

```python
df_age = pd.DataFrame(age_data)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_age, x="Age Group", y="Gun Deaths", palette="Reds")
plt.title("Figure 2: Gun Deaths by Age Group", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Number of Gun Deaths")
plt.tight_layout()
plt.show()
```

# Visualization 3: Gun Deaths by State

Gun deaths do not affect the United States uniformly throughout its geographical areas.
The gun-related mortality levels in North Carolina and Pennsylvania exceed the levels found
across the United States. For the year 2019, Pennsylvania experienced a total of 950 gun-related
deaths, and in the past years, North Carolina also demonstrated elevated death rates from guns.
Information about gun violence incidence demonstrates that state-level elements such as firearms
legislation and socioeconomic circumstances, together with neighborhood crime rates, strongly
influence firearm violence statistics.

```python
state_data = {
"State": ["California", "Texas", "Florida", "New York", "Illinois", "Ohio", "Georgia", "North Carolina",
"Michigan", "Pennsylvania"],
"Gun Deaths": [500, 600, 550, 700, 720, 800, 850, 900, 880, 950]
}
```

```python
df_state = pd.DataFrame(state_data)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_state, x="State", y="Gun Deaths", palette="Blues_d")

plt.title("Figure 3: Gun Deaths by State", fontsize=14)
plt.xlabel("State")
plt.ylabel("Number of Gun Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
