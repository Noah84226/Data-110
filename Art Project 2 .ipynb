I decided to use a multicolored tree for my art project to represent the flexibility of data and how code can change it. The many results that can be obtained from the same dataset when processed using various methods or methods are represented by each color on the tree. The tree roots' branching structure demonstrates how even little changes in code can result in completely new forms, while the bright colors demonstrate how data may be visually interpreted. This illustrates the notion that data is dynamic and serves as a basis for an infinite number of visual and analytical opportunities.

The structure of the tree also highlights the idea of branching logic in coding, which states that depending on the choices made, a single beginning point can result in a huge network of results. My artwork demonstrates that there is space for creativity, exploration, and surprise even in organized data by using a tree as a metaphor. It serves as a reminder that creativity and inventiveness are just as important to technology as accuracy.
